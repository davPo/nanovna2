# First comparisons between Keisight PNA and Nanovna v2 
406 MHz cavity filter
DPo - 8-2020


```python
import skrf as rf
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
rf.stylely()
plt.rcParams["figure.figsize"] = (10,10)
```

## S parameters loading

```python
pna_406 = rf.Network('filter406_pna.s2p') # measurement with Keysight PNA - Ecal module calibration  
nanovna_406_def = rf.Network('filter406_nanovna_defaultcal.s2p') # NanoVNA v2 with delivered standards - ideal
nanovna_406_real = rf.Network('filter406_nanovna_realcal.s2p') # NanoVNA v2 with delivered standards - models from s2p
# Note to myself Thru element APC3.5 needs to be de-embededed 
```

## S11 plotting

```python
plt.title('406 MHz filter $S_{11}$')
pna_406.s11.plot_s_smith(label='PNA')                    
nanovna_406_def.s11.plot_s_smith(label='NanoVNA default cal')                    
nanovna_406_real.s11.plot_s_smith(label='NanoVNA real cal')                    
```

## S22 plotting* 

```python
plt.title('406 MHz filter $S_{22}$')
pna_406.s22.plot_s_smith(label='PNA')                    
nanovna_406_def.s22.plot_s_smith(label='NanoVNA default cal')                    
nanovna_406_real.s22.plot_s_smith(label='NanoVNA real cal')   
```

## S21 Plotting

```python
plt.title('406 MHz filter $S_{21}$')
pna_406.s21.plot_s_db(label='PNA')                    
nanovna_406_def.s21.plot_s_db(label='NanoVNA default cal')                    
nanovna_406_real.s21.plot_s_db(label='NanoVNA real cal')   
```
